Séparation des données

```python
X = df.copy().drop(columns=['id', 'diagnosis'])
y = df['diagnosis'].map({'B':0, "M":1})
```

```python
y.value_counts()
```

Features Engeniering

```python
import seaborn as sns
```

```python
X['per_radius'] = X['radius_mean'] + X['perimeter_mean'] + X['area_mean']

X['per_radius_se'] = X['radius_se'] + X['perimeter_se'] + X['area_se']

X['per_radius_worst'] = X['radius_worst'] + X['perimeter_worst'] + X['area_worst']
```

```python
X = X.drop(columns=['radius_mean', 'radius_se', 'radius_worst', 'perimeter_mean', 'perimeter_se', 'perimeter_worst', 'area_mean', 'area_se', 'area_worst'])

X['per_radius_worst_norm'] = X['per_radius'] + X['per_radius_worst']
```

```python
X = X.drop(columns=['per_radius', 'per_radius_worst'])
```

```python
X = X.drop(columns=['symmetry_worst', 'fractal_dimension_mean', 'smoothness_se', 'smoothness_worst', 'compactness_se', 'smoothness_mean', 'fractal_dimension_se', 'fractal_dimension_worst', 'symmetry_se', 'symmetry_mean', 'texture_se'])
```

Correlation Plot

```python
plt.figure(figsize=(20, 16))
sns.heatmap(X.corr(), annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.show()
```

```python
for i in X.columns:
    sns.boxplot(X[i])
    plt.show()
```

Séparation des données, Standardisation et et réajustments

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.7, random_state=13)
```

```python
smote = SMOTE(random_state =42)
X_resampled, y_resampled = smote.fit_resample(X_train,y_train)
```

```python
scaler = StandardScaler()

X_train_s = scaler.fit_transform(X_resampled)
X_test_s = scaler.transform(X_test)
```

Essayons differents models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn.metrics import confusion_matrix
from scipy.stats import randint
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import auc, precision_recall_curve, roc_curve, roc_auc_score, average_precision_score
```

```python
param_dist = {
    'n_estimators': randint(50, 300),           # 50 à 300 arbres
    'max_depth': [5, 10, 20, None],             # Contrôle la complexité
    'min_samples_split': randint(2, 20),        # Évite le surapprentissage
    'min_samples_leaf': randint(1, 10),         # Évite les feuilles trop petites
    'max_features': ['sqrt', 'log2', 0.8],      # Diversité des arbres
    'class_weight': [None, 'balanced', {0:1, 1:5}]  # Ponde la classe minoritaire
}

# Créer le modèle
rf = RandomForestClassifier(random_state=42)

# Lancer la recherche (50 itérations pour rapidité)
random_search = RandomizedSearchCV(
    rf, 
    param_distributions=param_dist,
    n_iter=50,
    cv=5,
    scoring='average_precision',  # Optimise l'AP (adapté au déséquilibre)
    n_jobs=-1,                    # Utilise tous les cœurs CPU
    random_state=42
)

random_search.fit(X_train_s, y_resampled)

print("Meilleurs hyperparamètres :", random_search.best_params_)
```

```python
rf = RandomForestClassifier(
    n_estimators = random_search.best_params_['n_estimators'],
    max_depth = random_search.best_params_['max_depth'],
    min_samples_split = random_search.best_params_['min_samples_split'],
    min_samples_leaf = random_search.best_params_['min_samples_leaf'],
    max_features = random_search.best_params_['max_features'],
    class_weight = random_search.best_params_['class_weight']
)

rf.fit(X_train_s, y_resampled)
```

```python
y_pred = random_search.predict(X_test_s)
y_score = random_search.predict_proba(X_test_s)[:, 1]

roc_auc = roc_auc_score(y_test, y_score)
avg_precision = average_precision_score(y_test, y_score)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print(f"ROC AUC Score: {roc_auc:.2f}")
print(f"Average Precision Score: {avg_precision:.2f}")
```

Nous avons des très bon resultts avec un Random Forest

```python
feature_importances = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf.feature_importances_
}).sort_values(by="Importance", ascending=False)

print(feature_importances)
```

```python
plt.figure(figsize=(10, 5))
plt.barh(feature_importances["Feature"], feature_importances["Importance"], color='skyblue')
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.title("Feature Importance - RandomForestClassifier")
plt.gca().invert_yaxis()  # Pour afficher les features les plus importantes en haut
plt.show()
```

Essayons un autre model 

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import LogisticRegression

param_grid = {
    'penalty': ['l1', 'l2'],  # Retirez 'elasticnet' si inutile
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'saga'],  # 'liblinear' pour l1/l2, 'saga' pour elasticnet
    'class_weight': [None, 'balanced'],
    'max_iter': [5000],  # Augmentez le nombre d'itérations
    'tol': [1e-4]  # Tolérance pour arrêter plus tôt
}

# Création du modèle
model = LogisticRegression()

grid = GridSearchCV(model, param_grid=param_grid, cv=10, n_jobs=-1)

grid.fit(X_train_s, y_resampled)


print(f"Meilleurs hyperparamètres pour le SVC : {grid.best_params_}")
```

```python
lr = LogisticRegression(
    C = grid.best_params_['C'],
    penalty=grid.best_params_['penalty'],
    solver=grid.best_params_['solver'],
    class_weight = grid.best_params_['class_weight'],
    max_iter = grid.best_params_['max_iter'],
    tol = grid.best_params_['tol']
)

lr.fit(X_train_s, y_resampled)
```

```python
y_pred = lr.predict(X_test_s)
y_score = lr.predict_proba(X_test_s)[:, 1]

roc_auc = roc_auc_score(y_test, y_score)
avg_precision = average_precision_score(y_test, y_score)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print(f"ROC AUC Score: {roc_auc:.2f}")
print(f"Average Precision Score: {avg_precision:.2f}")
```

Comme nous pouvons voir nous avons un encore meuilleur score avec une Regression Linéaire

```python
y_scores = lr.decision_function(X_test_s)  

precision, recall, thresholds = precision_recall_curve(y_test, y_scores)

# Calculer l'AUC de la Precision-Recall curve
pr_auc = auc(recall, precision)

# Récupérer le seuil de base utilisé par le modèle SVC
threshold_base = 0  # Par défaut, SVC utilise un seuil de 0 sur la décision fonction
threshold_index = np.argmin(np.abs(thresholds - threshold_base))  # Trouver l'index du seuil le plus proche de 0

# Récupérer les valeurs de précision et rappel associées
recall_base = recall[threshold_index]
precision_base = precision[threshold_index]

# Tracer la Precision-Recall curve
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, color='darkorange', lw=2, label=f'Precision-Recall curve (AUC = {pr_auc:.2f})')

# Ajouter un cercle autour du point correspondant au seuil de base
plt.scatter(recall_base, precision_base, color='none', edgecolor='black', s=100, label="Threshold Base", linewidth=2)

# Ajouter les labels
plt.xlabel('Rappel (Recall)')
plt.ylabel('Précision (Precision)')
plt.title('Precision Recall Curve')
plt.legend(loc='lower left')
plt.show()
```

```python
# Calculer la courbe ROC
fpr, tpr, thresholds = roc_curve(y_test, y_scores)
roc_auc = auc(fpr, tpr)

# Récupérer le seuil de base utilisé par le modèle SVC (0 par défaut)
threshold_base = 0  
threshold_index = np.argmin(np.abs(thresholds - threshold_base))  # Trouver l'index du seuil proche de 0

# Récupérer les valeurs TPR et FPR associées
fpr_base = fpr[threshold_index]
tpr_base = tpr[threshold_index]

# Afficher la courbe ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'Courbe ROC (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label="Aléatoire (AUC = 0.5)")

# Ajouter un cercle autour du point correspondant au seuil de base
plt.scatter(fpr_base, tpr_base, color='none', edgecolor='black', s=100, label="Threshold Base", linewidth=2)

# Ajouter les labels
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taux de Faux Positifs (FPR)')
plt.ylabel('Taux de Vrais Positifs (TPR)')
plt.title('Courbe ROC')
plt.legend(loc='lower right')
plt.show()
```

Essayons un dernier model

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='rbf')

param_dist_svc = {
    'C': np.logspace(-4, 4, 20),  # Plage de valeurs pour C
    'gamma': ['scale', 'auto'] + np.logspace(-4, 4, 20).tolist()
}

grid = GridSearchCV(svc, param_grid=param_dist_svc, cv=10, n_jobs=-1)

grid.fit(X_train_s, y_resampled)

print(f"Meilleurs hyperparamètres pour le SVC : {grid.best_params_}")
```

```python
svc = SVC(
    kernel='rbf',
    C = grid.best_params_['C'],
    gamma = grid.best_params_['gamma'],
    probability=True
)

svc.fit(X_train_s, y_resampled)
```

```python
y_pred_svc = svc.predict(X_test_s)
y_score = svc.predict_proba(X_test_s)[:, 1]

roc_auc = roc_auc_score(y_test, y_score)
avg_precision = average_precision_score(y_test, y_score)

print("Accuracy:", accuracy_score(y_test, y_pred_svc))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_svc))
print("\nClassification Report:\n", classification_report(y_test, y_pred_svc))
print(f"ROC AUC Score: {roc_auc:.2f}")
print(f"Average Precision Score: {avg_precision:.2f}")
```

Comme nous pouvons le voir nous avons un tres bon resultat aussi avec un SVM
